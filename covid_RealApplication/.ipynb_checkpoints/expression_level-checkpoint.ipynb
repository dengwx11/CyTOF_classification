```r
markers <- rownames(seur)
celltypes <- colnames(A0)
rec_mat <- icu_mat <- as.data.frame(matrix(, ncol = 1 + length(markers)))
for(i in 1:length(rec_rst_files)) {
        
    rst <- readRDS(rec_rst_files[i])
    for(j in 1:length(celltypes)) {
        rec_mat <- rbind(rec_mat, c(celltypes[j], rst$W[,j]))
    }
}
rec_mat <- rec_mat[-1,]
colnames(rec_mat) <- c('cell_type', markers)
rec_mat
```

```r
rec_mat_long <- melt(rec_mat, id.vars = 'cell_type', variable.name = 'markers')
rec_mat_long$value <- as.numeric(rec_mat_long$value)
rec_mat_long$condition <- 'Recovery'
rec_mat_long
```

```r
icu_mat_long <- melt(icu_mat, id.vars = 'cell_type', variable.name = 'markers')
icu_mat_long$value <- as.numeric(icu_mat_long$value)
icu_mat_long$condition <- 'ICU'
icu_mat_long
```

```r
GeomSplitViolin <- ggproto(
  "GeomSplitViolin", 
  GeomViolin, 
  draw_group = function(self, data, ..., draw_quantiles = NULL) {
    data <- transform(data, 
                      xminv = x - violinwidth * (x - xmin), 
                      xmaxv = x + violinwidth * (xmax - x))
    grp <- data[1,'group']
    newdata <- plyr::arrange(
      transform(data, x = if(grp%%2==1) xminv else xmaxv), 
      if(grp%%2==1) y else -y
    )
    newdata <- rbind(newdata[1, ], newdata, newdata[nrow(newdata), ], newdata[1, ])
    newdata[c(1,nrow(newdata)-1,nrow(newdata)), 'x'] <- round(newdata[1, 'x']) 
    if (length(draw_quantiles) > 0 & !scales::zero_range(range(data$y))) {
      stopifnot(all(draw_quantiles >= 0), all(draw_quantiles <= 1))
      quantiles <- ggplot2:::create_quantile_segment_frame(data, draw_quantiles)
      aesthetics <- data[rep(1, nrow(quantiles)), setdiff(names(data), c("x", "y")), drop = FALSE]
      aesthetics$alpha <- rep(1, nrow(quantiles))
      both <- cbind(quantiles, aesthetics)
      quantile_grob <- GeomPath$draw_panel(both, ...)
      ggplot2:::ggname("geom_split_violin", 
                       grid::grobTree(GeomPolygon$draw_panel(newdata, ...), quantile_grob))
    } else {
      ggplot2:::ggname("geom_split_violin", GeomPolygon$draw_panel(newdata, ...))
    }
  }
)

geom_split_violin <- function (mapping = NULL, 
                               data = NULL, 
                               stat = "ydensity", 
                               position = "identity", ..., 
                               draw_quantiles = NULL, 
                               trim = TRUE, 
                               scale = "area", 
                               na.rm = FALSE, 
                               show.legend = NA, 
                               inherit.aes = TRUE) {
  layer(data = data, 
        mapping = mapping, 
        stat = stat, 
        geom = GeomSplitViolin, 
        position = position, 
        show.legend = show.legend, 
        inherit.aes = inherit.aes, 
        params = list(trim = trim, 
                      scale = scale, 
                      draw_quantiles = draw_quantiles, 
                      na.rm = na.rm, ...)
        )
}
```

```r
plot_lst <- list()
for(i in 1:length(markers)) {
    plot_lst[[i]] <- ggplot(mat_long[mat_long$markers == markers[i],],aes(x = cell_type, y = value, 
                                                                          fill = condition)) +
        geom_split_violin(trim=FALSE, alpha = 0.5, scale = 'width') +
      labs(x = 'Cell type', y = "Expression level", title =  markers[i]) +
      scale_fill_manual(values = c("#3D66CC", "#892110"))+ 
      theme(axis.text=element_text(size=12),
      axis.title=element_text(size=14,face="bold"),
      plot.title = element_text(size=14))
}
```
