```r
for(i in 1:length(seur_files_5)) {
    seur <- readRDS(seur_files_5[[i]])
    pred <- seur$pred
    label <- seur$label
    cm <- confusionMatrix(factor(seur$pred), factor(seur$label), dnn = c("Prediction", "Reference"))
    print(ggplot(as.data.frame(cm$table), aes(Prediction,Reference, fill= Freq)) +
            geom_tile() + geom_text(aes(label=Freq)) +
            scale_fill_gradient(low="white", high="#009194") +
            labs(x = "Reference",y = "Prediction") +
            theme(axis.text=element_text(size=12),
            axis.title=element_text(size=14,face="bold"),
            plot.title = element_text(size=14)))
}
```

```r
seur <- readRDS(seur_files_5[[1]])
pred <- seur$pred
label <- seur$label
cm <- confusionMatrix(factor(seur$pred), factor(seur$label), dnn = c("Prediction", "Reference"))
options(repr.plot.width = 8, repr.plot.height = 7, repr.plot.res = 300)
ggplot(as.data.frame(cm$table), aes(Prediction,Reference, fill= Freq)) +
            geom_tile() + geom_text(aes(label=Freq)) +
            ggtitle(seur$sample.id[1]) +
            scale_fill_gradient(low="white", high="#009194") +
            labs(x = "Reference",y = "Prediction") +
            theme(axis.text=element_text(size=12),
            axis.title=element_text(size=14,face="bold"),
            plot.title = element_text(size=14))
```
