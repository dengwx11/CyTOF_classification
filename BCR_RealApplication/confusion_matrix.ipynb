```r
library(caret)
library(Seurat)
```

```r
seur_files_5 <- list.files(path = '/gpfs/ysm/pi/zhao-data/wd262/new_cytof/write/BCR/5celltype', pattern='*seur*', full=T, ignore.case = TRUE)
seur_files_6 <- list.files(path = '/gpfs/ysm/pi/zhao-data/wd262/new_cytof/write/BCR/6celltype', pattern='*seur*', full=T, ignore.case = TRUE)
seur_files_7 <- list.files(path = '/gpfs/ysm/pi/zhao-data/wd262/new_cytof/write/BCR/7celltype', pattern='*seur*', full=T, ignore.case = TRUE)
```

```r
pred5 <- pred6 <- pred7 <- label5 <- label6 <- label7 <- c()
for(i in length(seur_files_5)) {
    seur5 <- readRDS(seur_files_5[[i]])
    seur6 <- readRDS(seur_files_6[[i]])
    seur7 <- readRDS(seur_files_7[[i]])
    pred5 <- c(pred5, seur5$pred)
    pred6 <- c(pred6, seur6$pred)
    pred7 <- c(pred7, seur7$pred)
    label5 <- c(label5, seur5$label)
    label6 <- c(label6, seur6$label)
    label7 <- c(label7, seur7$label)
}
```

```r
cm5 <- confusionMatrix(factor(seur5$pred), factor(seur5$label), dnn = c("Prediction", "Reference"))
ggplot(as.data.frame(cm5$table), aes(Prediction,Reference, fill= Freq)) +
        geom_tile() + geom_text(aes(label=Freq)) +
        scale_fill_gradient(low="white", high="#009194") +
        labs(x = "Reference",y = "Prediction") +
        theme(axis.text=element_text(size=12),
        axis.title=element_text(size=14,face="bold"),
        plot.title = element_text(size=14))
```

```r
cm6 <- confusionMatrix(factor(seur6$pred), factor(seur6$label), dnn = c("Prediction", "Reference"))
ggplot(as.data.frame(cm6$table), aes(Prediction,Reference, fill= Freq)) +
        geom_tile() + geom_text(aes(label=Freq)) +
        scale_fill_gradient(low="white", high="#009194") +
        labs(x = "Reference",y = "Prediction") +
        theme(axis.text=element_text(size=12),
        axis.title=element_text(size=14,face="bold"),
        plot.title = element_text(size=14))
```
